Analysis On Sleep Health & Lifestyle

Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Dataset & Explore it

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Sleep_health_and_lifestyle_dataset.csv")
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe(include='object'))
```

```python
print(df.describe())
```

```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna("No Disorder")
print(df.head())
```

```python
print(df.info())
```

```python
df['Sleep Disorder'].value_counts()
```

**Distribution of Sleep Disorder**

```python
# Count the occurrences of each category
disorder_counts = df['Sleep Disorder'].value_counts()

# Plot pie chart
plt.pie(disorder_counts, labels=disorder_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Sleep Disorders')
plt.show()
```

**Count of Persons in Each Age**

```python
sns.histplot(df['Age'], bins=10, kde=False, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**Age vs Sleep Cycle**

```python
plt.scatter(df['Age'], df['Sleep Duration'], color='teal', alpha=0.7)
plt.title('Age vs Sleep Cycle')
plt.xlabel('Age')
plt.ylabel('Sleep Cycle (Hours)')
plt.grid(True)
plt.show()
```

```python
correlation = df['Physical Activity Level'].corr(df['Stress Level'])
print("Correlation between physical activity and stress level:", correlation)
```

**Physical Activity vs Stress Level**

```python
plt.scatter(df['Physical Activity Level'], df['Stress Level'], color='coral', alpha=0.7)
plt.title('Physical Activity vs Stress Level')
plt.xlabel('Physical Activity (minutes/day)')
plt.ylabel('Stress Level')
plt.grid(True)
plt.show()
```

**Age vs BMI Category**

```python
plt.scatter(df['Age'], df['BMI Category'], color='teal', alpha=0.7)
plt.title('Age vs BMI Category')
plt.xlabel('Age')
plt.ylabel('BMI Category')
plt.grid(True)
plt.show()
```

**Average Heart rate by Gender**

```python
avg_hr = df.groupby('Gender')['Heart Rate'].mean()

avg_hr.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Heart Rate by Gender')
plt.ylabel('Average Heart Rate')
plt.xlabel('Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Sleep Disorder by Gender**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Sleep Disorder', data=df, palette='Set2', edgecolor="black")
plt.title('Gender vs Sleep Disorder')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Sleep Disorder')
plt.show()
```

**BMI Category by Gender**

```python
bmi_gender = df.groupby(['Gender', 'BMI Category']).size().unstack()

bmi_gender.plot(kind='bar', figsize=(8, 5), colormap='Accent', edgecolor="black")
plt.title('Gender vs BMI Category')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='BMI Category')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Occupation vs Sleep Disorder**

```python
occ_sleep = df.groupby(['Occupation', 'Sleep Disorder']).size().unstack()
occ_sleep.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set3', edgecolor="black")
plt.title('Occupation vs Sleep Disorder')
plt.xlabel('Occupation')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.legend(title='Sleep Disorder')
plt.tight_layout()
plt.show()
```

**Average Sleep Duration by each Occupation**

```python
# Step 1: Calculate average sleep duration per occupation
avg_sleep = df.groupby('Occupation')['Sleep Duration'].mean()

# Step 2: Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(avg_sleep.index, avg_sleep.values, color='orange')
plt.title('Average Sleep Duration by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Sleep Duration (hours)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
